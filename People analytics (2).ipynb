### Data Cleaning
In the case study it was identified that all the dates were entered 18 years in the past. We would adjust all of our relevant date fields due to the data issue identified by HR Analytica.
We will be incrementing all of the date fields except the arbitrary end date of 9999-01-01. 

```python
updated_employee = psql.sqldf('''
SELECT id,
            strftime('%Y/%m/%d',  birth_date, '+18 years')  as birth_date,
            first_name,
            last_name,
            gender,
            strftime('%Y/%m/%d', hire_date, '+18 years') as hire_date
            FROM employee
''')
updated_employee.head()
```

```python
updated_salary = psql.sqldf('''
SELECT employee_id,
amount,
strftime('%Y/%m/%d',  from_date, '+18 years')  as from_date,
strftime('%Y/%m/%d',  to_date, '+18 years')  as to_date
FROM salary
''')
updated_salary.head()
```

```python
updated_salary = psql.sqldf('''
SELECT employee_id,
amount,
strftime('%Y/%m/%d',  from_date, '+18 years')  as from_date,
case when strftime('%Y/%m/%d',  to_date) <> '9999/01/01' then strftime('%Y/%m/%d',  to_date, '+18 years')  else strftime('%Y/%m/%d',  to_date) end as to_date
FROM salary
''')
updated_salary.head()
```

## Background Recap
So far we have inspected the raw source data HR Analytica have shared with us inside the employees schema and we’ve applied corrections for a simple data input error by adding 18 years to each historic date record.


```python
# We create tenure for the data based on year end 31st December 2021
query1 = psql.sqldf('''
select employee_id, hire_date,
strftime('%Y/%m/%d', '2021-12-31T00:00:00.000Z') - hire_date as tenure

FROM updated_salary as u
join updated_employee as e
on e.id= u.employee_id
group by employee_id
order by employee_id asc
''')
query1.head()
```

```python
# SQL queries to calculate the average salary by Title
query2 = psql.sqldf('''
select title, round(avg(amount),2) as avg_amount_by_title
from (
select s.employee_id, t.title, s.amount
FROM updated_salary as s
join updated_title as t
on  s.employee_id=t.employee_id
and s.from_date=t.from_date
and s.to_date=t.to_date) as a
group by title
order by title desc
''')
query2.head(6)
```

```python
# SQL queries to calculate the average salary by department
query3 = psql.sqldf('''
with cte as (select a.employee_id, a.dept_name, s.amount
from (
select employee_id,department_id, from_date, to_date, dept_name
FROM updated_department_employee as de
join department as d
on d.id=de.department_id) as a
join updated_salary as s
on  s.employee_id=a.employee_id
and s.from_date=a.from_date
and s.to_date=a.to_date
)
select  dept_name, round(avg(amount),2) as avg_amount_by_department
from cte
group by dept_name
''')
query3.head(9)
```

```python
# What is the full name of the employee with the highest salary?
query4 = psql.sqldf('''
select first_name, last_name
from updated_employee as e
join (select employee_id
      from updated_salary as u
      where amount = (select max(amount)
                      from updated_salary as u)) as t
on e.id=t.employee_id
''')
query4.head()
```

```python
#Count of employees by tenure?
query5 = psql.sqldf('''
with cte as (
select employee_id, hire_date,
strftime('%Y/%m/%d', '2021-12-31T00:00:00.000Z') - hire_date as tenure

FROM updated_salary as u
join updated_employee as e
on e.id= u.employee_id
group by employee_id
order by employee_id asc)

select tenure, count(distinct employee_id) as tenure_count
from cte
group by tenure
order by tenure desc
''')
query5.head()
```

```python
#Which department has the least number of current employees?
query6 = psql.sqldf('''
select department_id, dept_name, count(distinct employee_id) as department_count
FROM updated_department_employee as de
join department as d
on d.id=de.department_id
where de.to_date='9999/01/01'
group by department_id, dept_name
order by department_count asc
limit 1
''')
query6.head()
```

```python
#What is the largest difference between minimimum and maximum salary values for all current employees?
query7 = psql.sqldf('''
select (max(amount)-min(amount)) as diff_max_min_salary
FROM updated_salary

''')
query7.head()
```

```python
#How many male employees are above the overall average salary value for the `Production` department?
query8 = psql.sqldf('''
with cte as (select s.employee_id,amount
FROM updated_salary as s
join (select employee_id, max(from_date) as from_date
FROM updated_salary
group by employee_id) as t
on t.employee_id=s.employee_id
and s.from_date=t.from_date),

ct as (select *
FROM updated_department_employee as de
where department_id=(select id
FROM  department as d
where dept_name="Production")),

c as (select cte.employee_id, amount, gender, department_id
from cte
join updated_employee as e
on e.id=cte.employee_id
join ct
on ct.employee_id=cte.employee_id)


select count(distinct employee_id) as male_production_employees
from c
where gender="M"
and amount > (select avg(amount) from c)


''')
query8.head()
```

```python
#Which title has the highest average salary for male employees?
query9 = psql.sqldf('''
with cte as (select s.employee_id,amount, s.from_date
FROM updated_salary as s
join (select employee_id, max(from_date) as from_date
FROM updated_salary
group by employee_id) as t
on t.employee_id=s.employee_id
and s.from_date=t.from_date)

select title, avg(amount) as average_salary_for_male_employees
from (select t.employee_id, title, amount, gender
    FROM updated_title as t
    join cte 
    on cte.employee_id=t.employee_id
    and cte.from_date=t.from_date
    join updated_employee as e
    on e.id=t.employee_id) as a
where gender="M"
group by title
order by average_salary_for_male_employees desc
limit 1

''')
query9.head()
```

```python
#Which title has the highest average salary for female employees?
query10 = psql.sqldf('''
with cte as (select s.employee_id,amount, s.from_date
FROM updated_salary as s
join (select employee_id, max(from_date) as from_date
FROM updated_salary
group by employee_id) as t
on t.employee_id=s.employee_id
and s.from_date=t.from_date)

select dept_name, avg(amount) as average_salary_for_female_employees
from (select de.employee_id, dept_name, amount, gender
    FROM updated_department_employee as de
    join cte 
    on cte.employee_id=de.employee_id
    and cte.from_date=de.from_date
    join updated_employee as e
    on e.id=de.employee_id
    join department as d
    on d.id=de.department_id
    ) as a
where gender="F"
group by dept_name
order by average_salary_for_female_employees desc
limit 1

''')
query10.head()
```

```python
#Which department has the most female employees?
query11 = psql.sqldf('''
select dept_name, count(distinct employee_id ) as female_employees
from (select de.employee_id, dept_name, gender
    FROM updated_department_employee as de
    join updated_employee as e
    on e.id=de.employee_id
    join department as d
    on d.id=de.department_id
    where gender="F") as a
group by dept_name
order by female_employees desc
limit 1
''')
query11.head(9)
```
